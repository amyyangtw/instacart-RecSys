```python
import pandas as pd
```

```python
order_ = pd.read_csv('./dataset/orders.csv')
order_ = order_[order_['eval_set'] == 'prior']
order_ = order_.iloc[:, 0:2]
order_
```

```python
item_ = pd.read_csv('./dataset/order_products__prior.csv')
order_item_ = pd.merge(order_, item_ , on='order_id', how='left')
order_item_ = order_item_.drop('add_to_cart_order', 1)
order_item_ = order_item_.iloc[:, 1:]
order_item_
```

```python
df = order_item_.groupby(['user_id', 'product_id'])['product_id'].size().reset_index(name='count')
df
```

```python
df.to_csv('./dataset/reorder_counts.csv', index=False)
```
